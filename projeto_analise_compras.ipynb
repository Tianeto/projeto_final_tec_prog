```python
# 1. O nível de consciencia ambiental aumenta ou diminui de acordo com a idade e renda?

#verificando a faixa etária do dataset
compras['age'].sort_values().unique()

#definindo 3 faixas de idade 
# Faixa 1 - 18 a 35 anos
# Faixa 2 - 36 a 55 anos
# Faixa 3 - 56 a 79 anos
faixa1 = compras['age'] <= 35
faixa2 = (compras['age'] >= 36) & (compras['age'] <= 55)
faixa3 = (compras['age'] >= 56) & (compras['age'] <= 79)

#Calculando a media do nível de consciencia ambiental por faixa etária
media_consc_ambiental_f1 = compras[faixa1]['environmental_awareness'].mean().round(2)
media_consc_ambiental_f2 = compras[faixa2]['environmental_awareness'].mean().round(2)
media_consc_ambiental_f3 = compras[faixa3]['environmental_awareness'].mean().round(2)
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_consc_ambiental_f1}')
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_consc_ambiental_f2}')
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_consc_ambiental_f3}')

# Verificando a faixa de renda do dataset
compras['monthly_income'].describe()

# Definindo as 3 faixas de renda
baixa_renda = compras['monthly_income'] <= 72450
media_renda = (compras['monthly_income'] > 72450) & (compras['monthly_income'] <= 190505)
alta_renda = compras['monthly_income'] > 190505

# Calculando a media do nível de consciencia ambiental por faixa etária
media_baixa_renda = compras[baixa_renda]['environmental_awareness'].mean().round(2)
media_media_renda = compras[media_renda]['environmental_awareness'].mean().round(2)
media_alta_renda = compras[alta_renda]['environmental_awareness'].mean().round(2)
print('=-' * 35)
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_baixa_renda}')
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_media_renda}')
print(f'Média de nível de consciencia ambiental para pessoas de baixa renda é {media_alta_renda}')

```

Resposta: Conforme verificamos, não há variação considerável no nível de consciencia ambiental nas
       diferentes faixas de renda e de idade.

```python
# 2. A quantidade de anos de uso do smartphone influencia na quantidade de compras online?


# Agrupando por anos de uso e calculando a média de gastos para cada ano
media_gasto_online_store = compras.groupby('smartphone_usage_years')[['avg_online_spend', 'avg_store_spend']].mean().round(2).reset_index()


# Verificando os resultados
display(media_gasto_online_store.sort_index())

```

Resposta: Olhando para a média de gastos, seja online ou em lojas físicas, vemos que a quantidade de anos de uso de smartphone não tem influência.

```python
# 3. Quem passa mais horas nas redes sociais tem um gasto maior em lojas fisicas ou online?

'''Nesta consulta, foi necessário fazer o arredontamento das horas, pois, sem isso, a coluna "daily_internet_hours" 
ficaria muito extensa e prejucaria a visualização dos dados'''

m_gastos_online_store = compras.groupby(compras ['daily_internet_hours'].round().astype(int))[['avg_online_spend', 'avg_store_spend']].mean().round(2).reset_index()


display(m_gastos_online_store)


```

Resposta: Não foi verificada relação entre o tempo que a pessoa navega na internet e o volume de gastos, seja online ou em lojas físicas.

```python
# 4. o nível de cohecimento tecnológico interfere no nível de confiança nos meios de pagamento online?.

tecnol_pgto_digital = (compras.groupby('tech_savvy_score')['online_payment_trust_score'].mean().round(2).reset_index()
.rename(columns={'tech_savvy_score': 'Nível de Conhecimento Tecnológico',
                 'online_payment_trust_score': 'Média de Confiança no Pagamento'}))

display(tecnol_pgto_digital)


```

Resposta: Analisando os diferentes níveis de conhecimento tecnológico, vemos que não há interferência na confiança nos meios de pagamento digital

```python
# 5. Qual a média de idade das pessoas que tem o maior volume de compras online?


# Definindo 5 faixas de idade 
#Faixa1 = 'age' <= 27 
#Faixa2 = 'age' >= 28 & 'age' <= 37 
#Faixa3 = 'age' >= 38 & 'age' <= 47  
#Faixa4 = 'age' >= 48 & 'age' <= 57 
#Faixa5 = 'age' >= 58 & 'age' <= 67 
#Faixa6 = 'age' >= 68

m_idade_x_comprasf1 = compras[compras['age']<= 27]['avg_online_spend'].mean().round(2)
m_idade_x_comprasf2 = compras[(compras['age']>= 28) & (compras['age']<= 37)]['avg_online_spend'].mean().round(2)
m_idade_x_comprasf3 = compras[(compras['age']>= 38) & (compras['age']<= 47)]['avg_online_spend'].mean().round(2)
m_idade_x_comprasf4 = compras[(compras['age']>= 48) & (compras['age']<= 57)]['avg_online_spend'].mean().round(2)
m_idade_x_comprasf5 = compras[(compras['age']>= 58) & (compras['age']<= 67)]['avg_online_spend'].mean().round(2)
m_idade_x_comprasf6 = compras[(compras['age']>= 68)]['avg_online_spend'].mean().round(2)

media_por_faixa_etaria = [['18 a 27', m_idade_x_comprasf1], ['28 a 37', m_idade_x_comprasf2],
['38 a 47', m_idade_x_comprasf3],['48 a 57', m_idade_x_comprasf4],['58 a 67', m_idade_x_comprasf5],
['67 a 79', m_idade_x_comprasf6]]


df_gasto_faixa_etaria = pd.DataFrame(media_por_faixa_etaria, columns=['Faixa Etária', 'Média compras Online'])
display(df_gasto_faixa_etaria)

```

Resposta: Com base nas informações da media de gastos com compras online por faixa etária, percebe-se um discreto aumento nas faixas acima de 58 anos.
